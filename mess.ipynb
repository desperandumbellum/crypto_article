```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eta_set = [0.2, 0.5, 0.99]
alpha_set = [0.2, 0.5]
l_set = np.arange(1, 501, 1)


def epsilon(l, aplha, eta, mu = 4, P_e = 8.5*10**(-7)):
    return (mu * eta * 10**( -alpha*l/10)) / (mu * eta * 10**( -alpha*l/10) + P_e)

fig = plt.figure(figsize=(16,9))
plt.rcParams.update({'font.size': 20})
for alpha in alpha_set:
    for eta in eta_set:
        epsilon_set = []
        for l in l_set:
            epsilon_set.append(epsilon(l, alpha, eta))
        plt.plot(epsilon_set, l_set, label = '$\\alpha = {}, \eta = {}$'.format(alpha, eta))
plt.legend()
plt.grid()
plt.ylabel('$\mathcal{l}$')
plt.xlabel('$\\varepsilon$')
plt.show()
```
